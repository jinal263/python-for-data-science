## 📖 Why Do We Need Loops?

Imagine you need to print "Hello" 5 times. Without loops:

```python
print("Hello")
print("Hello")
print("Hello")
print("Hello")
print("Hello")
```

What if you needed to print it 1000 times? Writing 1000 lines would be absurd!

**Loops** let you repeat code automatically. They are one of the most powerful concepts in programming.

> **Analogy:** A loop is like a washing machine cycle — it keeps repeating the same steps (wash, rinse, spin) until a condition is met (the timer runs out or the clothes are clean).

## 📖 The `for` Loop — Counting and Iterating

A `for` loop repeats code **for each item** in a sequence.

```python
for item in sequence:
    # This code runs once for each item
    do_something(item)
```

The most common way to use a for loop is with `range()`:

````python
# Print "Hello" 5 times — the easy way!
for i in range(5):
    print("Hello! This is repetition number", i)
````

````python
info_box("range() Starts at 0!",
    "<code>range(5)</code> produces: 0, 1, 2, 3, 4 (five numbers, starting from 0)<br>"
    "This is called <b>zero-based indexing</b> — Python counts from 0, not 1.<br>"
    "Think of it as: range(5) means '5 repetitions' but starting from 0.")
````

````python
# range() with different arguments
# range(stop) — counts from 0 to stop-1
print("range(5):", list(range(5)))           # [0, 1, 2, 3, 4]

# range(start, stop) — counts from start to stop-1
print("range(1, 6):", list(range(1, 6)))     # [1, 2, 3, 4, 5]

# range(start, stop, step) — counts with a specific step
print("range(0, 10, 2):", list(range(0, 10, 2)))    # [0, 2, 4, 6, 8]
print("range(10, 0, -1):", list(range(10, 0, -1)))  # [10, 9, 8, ..., 1]
````

````python
# Counting from 1 to 10
print("Counting from 1 to 10:")
for number in range(1, 11):   # 11 because the stop value is NOT included
    print(number, end=" ")    # end=" " prints on the same line
print()  # New line at the end
````

````python
# Multiplication table
number = 7
print(f"Multiplication table for {number}:")
print("-" * 20)
for i in range(1, 11):
    result = number * i
    print(f"{number} × {i} = {result}")
````

### Iterating Over Strings and Lists

`for` loops can iterate over any sequence, not just ranges:

````python
# Iterating over a string — one character at a time
name = "Python"
print("Letters in 'Python':")
for letter in name:
    print(letter, end=" → ")
print("done!")
````

````python
# Iterating over a list
fruits = ["apple", "banana", "cherry", "date"]
print("My favourite fruits:")
for fruit in fruits:
    print("  •", fruit)
````

````python
# Useful pattern: accumulating a total
numbers = [10, 20, 30, 40, 50]
total = 0  # Start with 0

for num in numbers:
    total = total + num    # Add each number to the total
    print(f"  Added {num}, running total = {total}")

print(f"Final total: {total}")
````

## 📖 The `while` Loop — Repeat Until Done

A `while` loop repeats code **as long as a condition is True**.

```python
while condition:
    # This code repeats as long as condition is True
    do_something()
```

> **Analogy:**
> - A `for` loop is like running **laps around a track** — you know how many laps in advance
> - A `while` loop is like **eating until you're full** — you don't know how many bites in advance

````python
# Simple while loop — countdown
count = 5
print("Countdown:")
while count > 0:
    print(count, end="... ")
    count -= 1         # CRITICAL: We must change the condition variable!
print("🚀 Liftoff!")
````

````python
warning_box("Infinite Loops!",
    "If the condition NEVER becomes False, the loop runs FOREVER! This is called an "
    "<b>infinite loop</b>.<br><br>"
    "Always make sure something inside the loop eventually makes the condition False.<br><br>"
    "If you accidentally create one in Jupyter: click <b>Kernel → Interrupt</b> "
    "or press the <b>⬛ Stop</b> button.")
````

````python
# While loop with user input — keep asking until they say 'quit'
print("Type messages (type 'quit' to stop):")
message = ""
while message != "quit":
    message = input("You: ")
    if message != "quit":
        print(f"Echo: {message}")
print("Goodbye!")
````

## 📖 `break` and `continue` — Loop Control

Sometimes you need to **exit a loop early** or **skip certain iterations**:

| Keyword | What it does | Analogy |
|---------|-------------|---------|
| `break` | Immediately exits the entire loop | Emergency exit — you leave the building |
| `continue` | Skips the rest of THIS iteration and moves to the next | Skipping a song — you move to the next track |

````python
# break — exit the loop early
print("Searching for the number 7:")
for num in range(1, 20):
    print(f"  Checking {num}...", end="")
    if num == 7:
        print(" Found it! 🎯")
        break   # Exit the loop immediately
    print(" not here")
print("Search complete.")
````

````python
# continue — skip certain iterations
print("Odd numbers from 1 to 10:")
for num in range(1, 11):
    if num % 2 == 0:    # If the number is even...
        continue        # ...skip it and go to the next number
    print(num, end=" ")
print()
````

## 🎯 Key Takeaways

1. **`for` loops** repeat for each item in a sequence — use when you know how many times
2. **`range()`** generates number sequences: `range(5)` → 0,1,2,3,4
3. **`while` loops** repeat as long as a condition is True — use when you don't know how many times
4. **`break`** exits a loop immediately; **`continue`** skips to the next iteration
5. Always ensure while loops can **terminate** — avoid infinite loops!
